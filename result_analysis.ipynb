```python
data_root = "./data/"

# list all the files in the data directory and for each file, load the json file into a dataframe
import os
files = os.listdir(data_root)
import json

dfs = []
for file in files:
    if file.endswith(".json"):
        data = json.load(open(data_root + file, encoding="UTF-8"))
        data = data['place_info']
        
        dfs = dfs + data
        
        
df = pd.DataFrame.from_records(dfs)
print(df.shape)
# remove any names that are duplicated
df = df.drop_duplicates(subset=['name'])
print(df.shape)
df
```

```python
df['rating_range'] = df['rating'].apply(lambda x: int(x))

def rating_record_range(x):
    if x < 10:
        return '0~10'
    elif x < 30:
        return '10~30'
    
    elif x < 50:
        return '30~50'
    
    else:
        return '50+'
    
    
df['rating_records_range'] = df['num_rating_records'].apply(lambda x: rating_record_range(x))
```

```python
# rearrange the columns as follows
# name, categories, restaurant type, address, rating, num_rating_records

df = df[['name', 'categories', 'restaurant type', 'address', 'rating', "rating_range", 'num_rating_records', "rating_records_range"]].copy()
df
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# 모집단 전체 평균 별점
overall_mean = (df['rating']*df['num_rating_records']).sum() / df['num_rating_records'].sum()

# 신뢰 구간과 p-value 계산 함수 정의
def calculate_confidence_interval_and_p_value(row, overall_mean):
    mean = row['rating']
    n = row['num_rating_records']
    
    # 별점의 표준 편차를 1.5로 가정 (1~5 사이 값에서의 분포)
    std_dev = 1.5
    
    # 표준 오차 계산
    se = std_dev / (np.sqrt(n) + 1e-8)
    
    # 95% 신뢰 구간 계산
    confidence_level = 0.95
    z_score = stats.norm.ppf((1 + confidence_level) / 2)
    margin_of_error = z_score * se
    confidence_interval = (mean - margin_of_error, mean + margin_of_error)
    
    # p-value 계산 (양측 검정)
    z = (mean - overall_mean) / se
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    return pd.Series({
        'ci_lower': confidence_interval[0],
        'ci_upper': confidence_interval[1],
        'p_value': round(p_value, 5)
    })

df = df[df['num_rating_records'] > 0].copy()
no_ratings = df[df['num_rating_records'] <= 0].copy()

# 각 음식점에 대해 신뢰 구간과 p-value 계산
df[['ci_lower', 'ci_upper', 'p_value']] = df.apply(calculate_confidence_interval_and_p_value, overall_mean=overall_mean, axis=1)


df = df.sort_values(by=['rating', 'p_value'], ascending = (False,True))

result = pd.concat([df, no_ratings])

# 결과 출력
result.head()
```
